```python
import pandas as pd

# サンプルデータの作成
data = {
    '国語': [10, 20, None],
    '数学': [40, None, 60]
}
dfA = pd.DataFrame(data)

# 欠損値を含む行を削除
dfB = dfA.dropna()

# 結果の表示
print(dfB)
```
